```python
import gymnasium as gym
import gymxq
```

```python
init_fen = "3k5/2P1P4/9/9/9/9/9/9/4p1p2/5K3 r - 100 0 190"
```

```python
env = gym.make(
    "xqv1",
    init_fen=init_fen,
    render_mode="ansi",
)
```

```python
for n in range(1, 10 + 1):
    print(f"{n}次模拟")
    env.reset()
    while True:
        action = env.sample_action()
        obs, r, t, tr, info = env.step(action)
        print(env.render())
        if t or tr:
            break
```
